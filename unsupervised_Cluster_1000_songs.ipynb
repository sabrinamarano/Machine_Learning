```python
test_0=cluster_0_energy.iloc[0,0:3].tolist()
test_0
```

```python
def euclidean_distance(centroid, song): 
    distance = (
        pd.DataFrame({
            'centroid':centroid,
            'song':song
        })
        .assign(dist_abs = lambda x: (x.centroid - x.song)**2)
        .agg({'dist_abs':'sum'})[0]**0.5
    )
    return distance

euclidean_distance(centroid=centroids_0, song=test_0)
```

```python
cluster_0_energy
distances=[]

for index, row in cluster_0_energy.iterrows(): 
    val= list(row)[:-1]
    distances.append(euclidean_distance(centroids_0, val))

cluster_0_energy=cluster_0_energy.assign(centroid_dist = distances)
```

```python
cluster_0_energy.sort_values(by='centroid_dist')
```
